## Kraken2 + bracken: for species level abundance estimates

Kraken2 assigns each read to various levels of taxonomic classification (eg some reads are only assigned to the "bacteria" domain...), and sometimes only one read is assigned to a species (the rest to upper classification levels because of some mismatches), it is tricky to interpret the raw results...
Enters bracken (Links to an external site.), a robust statistical software that scans the kraken2 reports and transforms the read level taxonomic assignments to estimations of species relative abundance in each sample.

parallel sbatch job arrays script:
