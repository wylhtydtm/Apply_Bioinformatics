## SQL bracken data explorations

```bash
module load r
sqlite3 -version
```

```bash
sqlite3 -cmd '.mode box' /shared/projects/form_2022_19/pascal/central_database/sample_collab.db 'select count(distinct(run_accession)) from bracken_abundances_long ;'
```

```bash
sqlite3 -cmd '.mode box' /shared/projects/form_2022_19/pascal/central_database/sample_collab.db 'select distinct(username) from (bracken_abundances_long abu left join sample_annot using(run_accession)) left join sample2bioinformatician s2b using(patient_code)';
```

```bash
sqlite3 -cmd '.mode box' /shared/projects/form_2022_19/pascal/central_database/sample_collab.db \
'select run_accession, taxon_name, new_est_reads reads,fraction_total_reads as frac,host_subject_id,nuc,host_body_site, host_disease_status,miscellaneous_parameter, Ct, total_reads from bracken_abundances_long abu left join sample_annot spl using(run_accession) limit 5;'
```

```bash
sqlite3 -cmd '.mode box' /shared/projects/form_2022_19/pascal/central_database/sample_collab.db \
'select run_accession, sum(fraction_total_reads) from bracken_abundances_long group by run_accession;'
```

```bash
sqlite3 -cmd '.mode box' /shared/projects/form_2022_19/pascal/central_database/sample_collab.db \
'select sum(new_est_reads) as total, taxon_name from bracken_abundances_long group by taxonomy_id order by total desc;'
```
